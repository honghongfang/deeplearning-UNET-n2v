```python
import numpy as np
from collections import OrderedDict
import tensorflow as tf
from PIL import Image
import tensorflow_probability as tfp
import matplotlib.pyplot as plt
from tifffile import imread
from tifffile import TiffWriter
```

```python
class MixingLayer(tf.keras.layers.Layer):
  def __init__(self, n_c, **kwargs):
    super().__init__(**kwargs)

    self.n_c = n_c
    self.M = None
    self.M2 = None

  def build(self, input_shape):
    d, h, w, n_f = input_shape[1], input_shape[2], input_shape[3], input_shape[4]
    
    self.M = self.add_weight(initializer=tf.keras.initializers.HeNormal(),shape=(n_f, self.n_c),name='m',dtype=tf.float32)
    self.M2 = self.M**2

    super().build(input_shape)

  def call(self, flr):
    mixed_flr = tf.matmul(flr, self.M2)
    return mixed_flr
```

```python
def loss_combined(masked_img, output, beta_1=1, beta_2=1):
  n_c = masked_img[-1]
  n_out = output[-1]  # n_out == n_c + n_f + n_c
  n_f = n_out - 2 * n_c

  denoised = output[..., :n_c]
  flr = output[..., n_c:n_c+n_f]
  mixed_flr = output[..., n_c+n_f:]


  loss1 = abs(masked_img - denoised)
  loss2 = beta_1*abs(mixed_flr - denoised)

  f_flat = tf.reshape(flr, shape=(-1, n_f))
  average_F = tf.math.reduce_mean(f_flat,axis=0,keepdims=True)
  F_centered = f_flat - average_F
  cov = tfp.stats.covariance(F_centered)
  mask_cov = (np.ones((n_f, n_f)) - np.eye((n_f))).astype(np.float32)
  mask_var = np.eye((n_f)).astype(np.float32)
  cov_masked = tf.multiply(cov, mask_cov)
  var_masked = tf.multiply(cov, mask_var)
  orthogonality_loss = tf.reduce_sum(tf.abs(cov_masked))
  unity_var_loss = tf.reduce_sum((var_masked-mask_var)**2)
  loss3 = beta_2 * (orthogonality_loss + unity_var_loss)
 
  
  tf.summary.scalar('loss_denoising', loss1)
  tf.summary.scalar('loss_rec', loss2)
  tf.summary.scalar('loss_orth', loss3)

  loss = loss1+loss2+loss3
  return loss
```

```python
class N2V(tf.keras.layers.Layer):
  def __init__(self, n_f, beta_1=1.,beta_2=1., **kwargs):
    super().__init__(**kwargs)

    self.beta_1 = beta_1
    self.beta_2 = beta_2
    self.n_f = n_f

    mask_cov = (np.ones((self.n_f, self.n_f)) - np.eye((self.n_f))).astype(np.float32)
    mask_var = np.eye((self.n_f)).astype(np.float32)

    self.mask_cov = tf.constant(mask_cov)
    self.mask_var = tf.constant(mask_var)

  def build(self, input_shape):

    d, h, w, n_c = input_shape[1], input_shape[2], input_shape[3], input_shape[4]
    self.denoised = self.add_weight(initializer=tf.keras.initializers.HeNormal() ,shape=(d, h, w, n_c/2-self.n_f),name='denoised',dtype=tf.float32)
    self.flr = self.add_weight(initializer=tf.keras.initializers.HeNormal(),shape=(d, h, w, self.n_f),name='flr',dtype=tf.float32)
    self.mixed_flr = self.add_weight(initializer=tf.keras.initializers.HeNormal(),shape=(d, h, w, n_c/2),name='mixed_flr',dtype=tf.float32)
    super().build(input_shape)

  def ort_loss(self):

    f_flat = tf.reshape(self.flr, shape=(-1, self.n_f))
    average_F = tf.math.reduce_mean(f_flat,axis=0,keepdims=True)
    F_centered = f_flat - average_F
    cov = tfp.stats.covariance(F_centered)

    cov_masked = tf.multiply(cov, self.mask_cov)
    var_masked = tf.multiply(cov, self.mask_var)
    orthogonality_loss = tf.reduce_sum(tf.abs(cov_masked))
    unity_var_loss = tf.reduce_sum((var_masked-self.mask_var)**2)

    return self.beta_2 * (orthogonality_loss + unity_var_loss)
  
  def MAE(a, b):
    return abs(a - b)

  def call(self, C):

    denoiseing_loss = MAE(masked_img,denoised)
    reconstruction_loss = self.beta_1 * MAE(mixed_flr,denoised)
    orthogonality_loss =  self.ort_loss()
    
    loss =  denoiseing_loss + reconstruction_loss + orthogonality_loss
    return loss
```
